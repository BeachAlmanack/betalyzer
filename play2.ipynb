```python
import Quandl
import pandas as pd
import numpy as np
import datetime
```

```python
idx = pd.date_range(start='01-01-2010', end='01-01-2016')
tickers = ['PIH', 'FLWS', 'FCCY', 'SRCE', 'VNET', 'TWOU', 'JOBS', 'CAFD', 'EGHT', 'AVHI']
dt = {t: np.random.uniform(-.05, .05, len(idx)) for t in tickers}
df = pd.DataFrame(data=dt, index=idx)

mkt = 'AVHI'
window = 200
```

```python
covs = df.rolling(window=window).cov(df[mkt], pairwise=True)
var = df[mkt].rolling(window=window).var()
beta = covs.div(var,axis=0)
```
